```python
import csv

r = input('2017~2019: ')

f = open('Daegu.csv',encoding = 'cp949')
data = csv.reader(f)
header = next(data)

bb = 0
cc = 0
dd = 0
rr = 0
for i in data:
    if r == i[0] and '사고건수' in i[1]:
        bb = int(i[2].replace(',',''))
        rr = int(i[0])
        print("%d년도 사고건수 %d"%(rr,bb))
        
    elif r == i[0] and '부상자수' in i[1]:
        cc = int(i[2].replace(',',''))
        rr = int(i[0])
        print("%d년도 부상자수 %d"%(rr,cc))
        
    elif r == i[0] and '사망건수' in i[1]:
        dd = int(i[2].replace(',',''))
        rr = int(i[0])
        print("%d년도 사망건수 %d"%(rr,dd))
print(type(bb))
print(type(cc))
print(type(dd))
f.close()

f = open('drunk.csv',encoding = 'cp949')
data1 = csv.reader(f)

tt = 0

for m in data1 :
    if m[0] == '대구광역시'  and r == '2017':
        tt = int(m[8].replace(',',''))
        rr = int(r)
        print("%d년도 음주운전건수 %d"%(rr,tt))
    elif m[0] == '대구광역시'  and r == '2018':
        tt = int(m[11].replace(',',''))
        rr = int(r)
        print("%d년도 음주운전건수 %d"%(rr,tt))
    elif m[0] == '대구광역시'  and r == '2019':
        tt = int(m[14].replace(',',''))
        rr = int(r)
        print("%d년도 음주운전건수 %d"%(rr,tt))



f.close()

import matplotlib.pyplot as plt
#plt.rc('font',family = 'NanumGothic')
plt.figure(figsize = (10,10))
size = [bb,cc,dd,tt]
label = ['qqq','qq2','qq3','qq4']
#plt.axis('equal')
plt.pie(size, labels = label, autopct = '%.1f%%',textprops={'fontsize':5},colors = 'colors')
plt.legend()
plt.show()

```
